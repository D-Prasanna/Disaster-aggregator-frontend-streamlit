```python
import geopandas as gpd

# Load the shapefile
world = gpd.read_file("C:\\Users\\prasa\\Hackathon\\testing stuff\\ne_110m_admin_0_countries\\ne_110m_admin_0_countries.shp")

# Example of how to merge your disaster data with the world map
import pandas as pd

# Example data
data = {
    'Country': ['United States', 'Canada', 'Mexico', 'China', 'India', 'Japan', 'Russia'],
    'DisasterCount': [1000, 500, 400, 600, 700, 800, 300]  # Example numbers
}

df = pd.DataFrame(data)

# Merge the shapefile with your data
world = world.merge(df, how="left", left_on="ADMIN", right_on="Country")

# Plot the map as described earlier
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1)
world.plot(column='DisasterCount', ax=ax, legend=True,
           legend_kwds={'label': "Number of Disaster-Related Events",
                        'orientation': "horizontal"},
           cmap='OrRd')

plt.title('World Map Heatmap of Disaster-Related Events')
plt.show()
```
